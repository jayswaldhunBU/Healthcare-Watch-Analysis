#Relationship between Flourishing Scale and Garmin Data

##Reading Garmin Data

```python
import pandas as pd
df = pd.read_csv("finalGarminData_6Month.csv")
```

```python
df.rename(columns={"Participant Id": "mbl_cod"},inplace = True)
```

```python
df = df.drop(columns=['new_date', 'date'])
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
```

```python
df
```

```python
df = df.drop(columns=['Date'])
df = df[df.Month != 4]
df = df[df.Month != 11]
```

```python
df
```

```python
n = len(df.mbl_cod.unique())
```

```python
n
```

##Reading Survey Data

```python
survey_df_2 = pd.read_excel('7105000 (2).xlsx')
```

```python
survey_df_2
```

```python
survey_df_2['local_time'] = [d.date() for d in survey_df_2['local_time']]
survey_df_2['Date'] = pd.to_datetime(survey_df_2['local_time'])
survey_df_2['Month'] = survey_df_2['Date'].dt.month
```

```python
survey_df_2
```

```python
survey_df_2 = survey_df_2.drop(columns=['Date'])
```

```python
survey_df_2
```

##Merging both data

```python
merge_df= df.merge(survey_df_2, on=['mbl_cod','Month'], how='inner')
```

```python
merge_df
```

## Flourishing Scale

```python
merge_df['Happiness_LifeSatisfaction'] = merge_df['HAPPINESS'] + merge_df['LIFE_SATISFACTION']
merge_df['Mental_PhysicalHealth'] = merge_df['MENTAL_HEALTH'] + merge_df['PHYSICAL_HEALTH']
merge_df['Meaning_Purpose'] = merge_df['PURPOSE'] + merge_df['WORTHWHILE']
merge_df['Character_Virtue'] = merge_df['PROMOTE_GOOD'] + merge_df['DELAYED_HAPPINESS']
merge_df['Social_Relationship'] = merge_df['CONTENT_RELATIONSHIPS'] + merge_df['SATISFYING_RELATIONSHIPS']
merge_df['Financial_MaterialStability'] = merge_df['LIVING_EXPENSES'] + merge_df['FOOD_HOUSING']
```

```python
merge_df['Flourishing_Scale'] = merge_df['Happiness_LifeSatisfaction'] + merge_df['Mental_PhysicalHealth'] + merge_df['Meaning_Purpose'] + merge_df['Character_Virtue'] + merge_df['Social_Relationship'] + merge_df['Financial_MaterialStability']
```

##Total number of participants 

```python
n = len(merge_df.mbl_cod.unique())
```

```python
n
```

##Grouping by month

```python
merge_df = merge_df.groupby(by="Month", as_index=False).mean()
```

```python
merge_df.columns
```

```python
merge_df
```

##Domain 1 - Mental and Physical Health

```python
import matplotlib.pyplot as plt
import seaborn as sns
color = 'tab:red'
fig, ax1 = plt.subplots()
ax1.plot(merge_df.Month, merge_df['calories'], color )
ax1.set_ylabel('calories', color=color ) 
ax1.tick_params(axis='y', labelcolor=color )
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.plot(merge_df.Month, merge_df['Mental_PhysicalHealth'], color)
ax2.set_ylabel('Mental_PhysicalHealth', color=color) 
ax2.tick_params(axis='y', labelcolor=color)
ax1.set_xlabel('Month') 
plt.show()
```

Interpretation

H0: correlation is not statistically significant.

H1: correlation is statistically significant.

```python
from scipy.stats.stats import pearsonr
```

```python
stat, p = pearsonr(merge_df['calories'], merge_df['Mental_PhysicalHealth'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('correlation is not statistically significant')
else:
 print('correlation is statistically significant')
```
