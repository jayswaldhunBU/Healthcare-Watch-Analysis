```python
import pandas as pd
df = pd.read_csv("finalGarminData_6Month.csv")
```

```python
df.rename(columns={"Participant Id": "mbl_cod"},inplace = True)
```

```python
df = df.drop(columns=['new_date', 'date'])
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
```

```python
df = df.drop(columns=['Date'])
df = df[df.Month != 4]
df = df[df.Month != 11]
```

```python
df
```

```python
computer_chklist_df_2 = pd.read_excel('7102000 (1).xlsx')
```

```python
computer_chklist_df_2['local_time'] = [d.date() for d in computer_chklist_df_2['local_time']]
computer_chklist_df_2['Date'] = pd.to_datetime(computer_chklist_df_2['local_time'])
computer_chklist_df_2['Month'] = computer_chklist_df_2['Date'].dt.month
```

```python
computer_chklist_df_2
```

```python
computer_chklist_df_2 = computer_chklist_df_2.drop(columns=['Date'])
```

```python
merge_df= df.merge(computer_chklist_df_2, on=['mbl_cod','Month'], how='inner')
```

```python
merge_df
```

```python
merge_df.columns
```

```python
new_cols = ['HORIZONTAL_THIGHS_WHEN_SEATED', 'VERTICAL_LOWER_LEGS','FEET_FLAT_ON_FLOOR_OR_FOOTREST', 'STRAIGHT_WRISTS', 'ADJUST_EASILY',
       'PADDED_ROUNDED', 'ADJUSTABLE_BACKREST', 'ADJUSTABLE_ARMRESTS',
       'LUMBAR_SUPPORT', 'CASTERS', 'KEYBOARD_SURFACE', 'KEYBOARD_TILT',
       'KEYBOARD_DETACHABLE', 'DOCUMENT_HOLDER', 'SCREEN_GLARE',
       'BRIGHTNESS_CONTRAST', 'VIEWING_DISTANCE', 'KNEE_FOOT_SPACE',
       'HANDED_ACTIVITIES', 'REST_BREAKS', 'JOB_ROTATION', 'TASK_PACE',
       'POSTURE_TRAINING', 'METHODS_TRAINING', 'WMSD_SIGNS',
       'WORKSTATION_ADJUSTMENT']
```

```python
merge_df['Total_Score'] = merge_df[new_cols].sum(axis=1)
```

```python
merge_df
```

```python
n = len(merge_df.mbl_cod.unique())
```

```python
n
```

```python
merge_df = merge_df.groupby(by="Month", as_index=False).mean()
```

```python
merge_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
color = 'tab:red'
fig, ax1 = plt.subplots()
ax1.plot(merge_df.Month, merge_df['calories'], color )
ax1.set_ylabel('calories', color=color ) 
ax1.tick_params(axis='y', labelcolor=color )
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.plot(merge_df.Month, merge_df['Total_Score'], color)
ax2.set_ylabel('Total_Score', color=color) 
ax2.tick_params(axis='y', labelcolor=color)
ax1.set_xlabel('Month') 
plt.show()
```

```python
from scipy.stats.stats import pearsonr
```

```python
stat, p = pearsonr(merge_df['calories'], merge_df['Total_Score'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('correlation is not statistically significant')
else:
 print('correlation is statistically significant')
```
