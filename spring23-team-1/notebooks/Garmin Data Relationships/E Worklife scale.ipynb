```python
import pandas as pd
df = pd.read_csv("finalGarminData_6Month.csv")
```

```python
df.rename(columns={"Participant Id": "mbl_cod"},inplace = True)
df = df.drop(columns=['new_date', 'date'])
df
```

```python
df['Date'] =  pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df
```

```python
df['Date'] = df['Date'].dt.date
```

```python
#df = df.drop(columns=['Date'])
df = df[df.Month != 4]
df = df[df.Month != 11]
df
```

```python
n = len(df.mbl_cod.unique())
```

# E-worklife scale

```python
merge_df['Organisational_Trust'] = merge_df['TRUST_1'] + merge_df['TRUST_1'] + merge_df['TRUST_2'] + merge_df['TRUST_3'] + merge_df['TRUST_4'] + merge_df['TRUST_5']
merge_df['Flexibility'] = merge_df['FLEXIBILITY_6'] + merge_df['FLEXIBILITY_7'] + merge_df['FLEXIBILITY_8'] + merge_df['FLEXIBILITY_9'] 
merge_df['Worklife_Interference'] = merge_df['WORK_LIFE_10'] + merge_df['WORK_LIFE_11'] + merge_df['WORK_LIFE_12'] + merge_df['WORK_LIFE_13'] + merge_df['WORK_LIFE_14'] + merge_df['WORK_LIFE_15']                              
merge_df['Effectiveness_Productive'] = merge_df['PRODUCTIVITY_16'] + merge_df['PRODUCTIVITY_17'] + merge_df['PRODUCTIVITY_18'] + merge_df['PRODUCTIVITY_19'] + merge_df['PRODUCTIVITY_20'] + merge_df['PRODUCTIVITY_21']
                                        
```

```python
merge_df['E_Worklife_Scale'] = merge_df['Organisational_Trust'] + merge_df['Flexibility'] + merge_df['Worklife_Interference'] + merge_df['Effectiveness_Productive']
```

```python
n = len(merge_df.mbl_cod.unique())
```

## Organisational Trust

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats.stats import pearsonr
```

```python
merge_df = merge_df.groupby(by="Month", as_index=False).mean()
```

```python
color = 'tab:red'
fig, ax1 = plt.subplots()
ax1.plot(merge_df.Month, merge_df['calories'], color , marker = 'o')
ax1.set_ylabel('calories', color=color ) 
ax1.tick_params(axis='y', labelcolor=color )
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.plot(merge_df.Month, merge_df['Organisational_Trust'], color, marker = 'o')
ax2.set_ylabel('Organisational_Trust', color=color) 
ax2.tick_params(axis='y', labelcolor=color)
ax1.set_xlabel('Month') 
plt.title('Organisational_Trust vs calories')
plt.show()

stat, p = pearsonr(merge_df['calories'], merge_df['Organisational_Trust'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('correlation is not statistically significant')
else:
    print('correlation is statistically significant')
```
