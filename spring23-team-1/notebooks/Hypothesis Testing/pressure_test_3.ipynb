```python
import pandas as pd
DA_df = pd.read_excel('Daily_AM.xlsx')
DP_df = pd.read_excel('Daily_PM.xlsx')
DE_df = pd.read_excel('Daily_END.xlsx')
FA_df = pd.read_excel('Friday_AM.xlsx')
FP_df = pd.read_excel('Friday_PM.xlsx')
FE_df = pd.read_excel('Friday_END.xlsx')
```

```python
DA_new_df=DA_df.groupby('LOCATION_AM').STRESS.mean()
DA_new_df
```

```python
import matplotlib.pyplot as plt

location = ['Bedroom', 'Home office room', 'Company office', 'Living room', 'Kitchen', 'Dinning room', 'Away from home']
plt.figure(figsize=(20,7))
plt.xlabel('Location')
plt.ylabel('Stress')
plt.title('Daily AM (3 Month)', fontsize = 30)
plt.bar(location, DA_new_df)
plt.savefig('Daily AM (3 Month)', dpi=300)
plt.show()
```

```python
FE_new_df=FE_df.groupby('LOCATION_END').STRESS.mean()
FE_new_df[3.0]=0.0
FE_new_df=FE_new_df.sort_index()
FE_new_df
```

```python
import matplotlib.pyplot as plt

location = ['Bedroom', 'Home office room','Company office', 'Living room', 'Kitchen', 'Dinning room', 'Away from home']
plt.figure(figsize=(20,7))
plt.xlabel('Location')
plt.ylabel('Stress')
plt.title('Friday END (3 Month)', fontsize = 30)
plt.bar(location, FE_new_df)
plt.savefig('Friday END (3 Month)', dpi=300)

plt.show()
```

```python
import matplotlib.pyplot as plt
for i in range(1,8):
    x=['AM', 'PM', 'END']
    y=[DA_new_df.loc[i], DP_new_df.loc[i], DE_new_df.loc[i]]
    location = ['Bedroom', 'Home office room','Company office', 'Living room', 'Kitchen', 'Dinning room', 'Away from home']
    label=location[i-1]
    plt.plot(x,y, label=label)
    plt.legend(loc='best', fontsize=8)
plt.xlabel('Time')
plt.ylabel('Stress')    
plt.title('Stress variation in Daily (3 Month)')  
plt.savefig('Stress variation in Daily (3 Month)', dpi=300)
plt.show()
```
