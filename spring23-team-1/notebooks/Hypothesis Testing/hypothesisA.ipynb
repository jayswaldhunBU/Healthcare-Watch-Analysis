## Hypothesis A 
## Participants’ age will negatively correlate with financial and material stability (the last two questions on the Flourishing Scale) 
### Its comparing Age and Living Expense Column

### 3 Month Data

```python
import pandas as pd

#Financial and Material Stability depicts -How often do you worry about being able to meet normal monthly living expenses?
#Friday AM data consists the Living Expense Column
df = pd.read_csv("FridayAM.csv")  
#Demographic Data consists the Age Column
df1 = pd.read_csv("Demographic Data - Sheet1.csv")
df.head()
df1.head()
```

```python
# For joining the 2 dataframes we need same name for id i.e mbl_cod
df1=df1.rename(columns={'Mobile ID':'mbl_cod'})
```

```python
#Joining the 2 dataframes
result = pd.merge(df,df1,on = 'mbl_cod')
result.head
```

```python
#Only selecting id,Age and Living_Expense Column
result = result[['mbl_cod','Age ','LIVING_EXPENSES']]
result['LIVING_EXPENSES'] = result['LIVING_EXPENSES']-1
result.dropna(inplace=True)
result
```

```python
import matplotlib.pyplot as plt
```

```python
result_3mnths = result.groupby('Age ')['LIVING_EXPENSES'].mean()
result_3mnths.plot(kind='line',title='Age of Respondant vs Average score of financial and material stability(3 months) ',
               ylabel='Average score', xlabel='Age', figsize=(8, 4))
plt.xticks(rotation=0)
```

### 6 Month Data

```python
df2 = pd.read_csv("FridayAm6Mnths.csv")
```

```python
result_6mnths = pd.merge(df1,df2,on = 'mbl_cod')
result_6mnths.head
```

```python
result_6mnths = result_6mnths[['mbl_cod','Age ','LIVING_EXPENSES']]
result_6mnths['LIVING_EXPENSES'] = result_6mnths['LIVING_EXPENSES']-1
result_6mnths.dropna(inplace=True)
result_6mnths
```
