```python
survey_df = pd.read_excel('7105000 (2).xlsx')
survey_df['local_time'] = [d.date() for d in survey_df['local_time']]
survey_df['Date'] = pd.to_datetime(survey_df['local_time'])
survey_df['Month'] = survey_df['Date'].dt.month
survey_df_abs = survey_df[['Month', 'LIFE_SATISFACTION', 'HAPPINESS', 'PHYSICAL_HEALTH', 'MENTAL_HEALTH']]
survey_df_abs = survey_df_abs.groupby(by="Month", as_index=False).mean()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df = df.groupby(by="Month", as_index=False).mean()
df = df[df.Month != 4]
df = df[df.Month != 11]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax1 = plt.subplots()
ax1.plot(df.Month, df['calories'], '.r-')
ax1.set_ylabel('calories', color='r') 
ax1.tick_params(axis='y', labelcolor='r')
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.plot(survey_df_abs.Month, survey_df_abs['PHYSICAL_HEALTH'], color)
ax2.set_ylabel('PHYSICAL_HEALTH', color=color) 
ax2.tick_params(axis='y', labelcolor=color)
plt.show()
```
